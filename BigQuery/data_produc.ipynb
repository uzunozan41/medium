```python
from google.cloud import bigquery
import pandas as pd
from random import randint
```

```python
client=bigquery.Client(project="medium-xxxx")
```

```python
df=pd.read_csv("xvericom_20221205075906.csv")
df
```

```python
first_names=df["first_name"].values
last_names=df["last_name"].values
```

```python
years=[2010,2011,2012,2012,2014,2015,2016,2017,2018,2019,2020,2021,2022]
books=["write fang","animal farm","the picture of dorian gray"
       ,"the lost world","frankenstein","dracula","the time machine"
      ,"peter pan","rebecca of sunnybrook farm","hans brinker or the silver skates"
      "tale of peter rabbit","anne of green gables","gulliver's travels"]
```

```python
def produce_data(id):
    data={}
    data["id"]=id
    data["first_name"]=first_names[randint(0,99)]
    data["last_name"]=last_names[randint(0,99)]
    over=randint(0,5)
    borrow_books=[]
    
    for i in range(0,over):
        borrow_book={}
        borrow_book["year"]=years[randint(0,12)]
        borrow_book["name"]=books[randint(0,11)]
        borrow_books.append(borrow_book)
        
    data["borrow_books"]=borrow_books
    
    
    return data
```

```python
produce_data(1)
```

```python
data_list=[]

for i in range(1,1000):
    data=produce_data(i)
    data_list.append(data)
    print(data)
```

```python
client.insert_rows_json(table="medium-xxxx.library.student_record",json_rows=data_list)
```
