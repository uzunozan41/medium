```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
gdf=gpd.read_file("export.geojson")
```

```python
gdf.head()
```

```python
gdf.columns
```

```python
gdf_simple=gdf[["id","name","type","geometry"]]
gdf_simple
```

```python
gdf_simple.to_file("export data/test.shp")
```

```python
gdf_simple["geo_type"]=gdf_simple["geometry"].geom_type
gdf_simple
```

```python
gdf_simple["geo_type"].drop_duplicates()
```

```python
gdf_polygon=gdf_simple[(gdf_simple["geo_type"]=="Polygon") | (gdf_simple["geo_type"]=="MultiPolygon")]
gdf_polygon
```

```python
gdf_point=gdf_simple[gdf_simple["geo_type"]=="Point"]
gdf_point
```

```python
gdf_line=gdf_simple[gdf_simple["geo_type"]=="LineString"]
gdf_line
```

```python
gdf_polygon.to_file("export data/test_polygon.shp")
gdf_line.to_file("export data/test_line.shp")
gdf_point.to_file("export data/test_point.shp")
```

```python
population_df=gpd.read_file("population_data.csv",encoding="utf-8")
population_df
```

```python
population_df=population_df.drop(["geometry"],axis=1)
population_df
```

```python
merge_df=gdf_polygon.merge(population_df,left_on="name",right_on="district")
merge_df
```

```python
merge_df["population"]=merge_df["population"].astype("int")
```

```python
fig,ax=plt.subplots(figsize=(25,15))

merge_df.boundary.plot(ax=ax,color="black")
merge_df.plot(column="population",ax=ax,legend=True,cmap="Blues")

ax.grid()
ax.set_title("Population By Kocaeli",fontsize=30)
ax.set_xlabel("medium.com/@uzunozan41",fontsize=20)
ax.set_ylabel("Geopandas IV (GeoJson) EN",fontsize=20)
```

```python
fig,(ax0,ax1,ax2)=plt.subplots(1,3,figsize=(20,5))

gdf_polygon.plot(ax=ax0)
gdf_line.plot(ax=ax1)
gdf_point.plot(ax=ax2)


ax0.set_title("Polygon",size=20)
ax1.set_title("Line",size=20)
ax2.set_title("Point",size=20)
```

```python
fig1,(ax01,ax11)=plt.subplots(1,2,figsize=(20,5))

df_istanbul=gpd.read_file("export data/istanbul.shp")
df_kocaeli=gpd.read_file("export data/Ankara.shp")

df_istanbul.plot(ax=ax01)
df_kocaeli.plot(ax=ax11)

ax01.set_title("İstanbul \n Multipolygon",size=20)
ax11.set_title(" Ankara \n Polygon",size=20)
```
